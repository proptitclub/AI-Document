# LeNet5 Convolution Neural Network 
-------------------


**INPUT => CONV + RELU => POOL => CONV + RELU => POOL => FC + RELU => FC => OUTPUT**

# SUMMARY OF LeNet

Nhận xét: 
* Đây là 1 kiến trúc mạng CNN đơn giản ra đời năm 1994 bởi Yann LeCun 
* 3 Convolution Layers
    * sử dụng hàm activation: sigmoid hoặc tanh dẫn đến **vanishing gradient** 
    * no padding
* 2 Pooling Layers xen giữa với stride = 2.
* 2 Fully-connected Layers và sử dụng hàm action: softmax để dự đoán đầu ra. ( Bảng trên thiếu 1 FC Layer )
* 1 Output Layer có 10 nodes.


MẠNG LeNet5 đã cải thiện hơn so với LeNet1, ACCURACCY trên bộ data MNIST đạt 0.99% khi dùng Lenet5. 

Vì thế, Lenet5 đã phổ biến hơn và khi nói đến LeNet thì nói tới kiến trúc mạng LeNet5.

Sự khác nhau giữa LeNet1 và Lenet5 đó chính là feater map 
* LeNet1: 4 feature map tại Conv Layer đầu tiên và ko có FC Layers.
* LeNet4: 4 feature map tại Conv Layer đầu tiên và có 1 FC Layers.
* LeNet5: 6 feature map tại Conv Layer đầu tiên và có 2 FC Layers.
