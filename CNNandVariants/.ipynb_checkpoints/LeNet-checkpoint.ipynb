# LeNet5 Convolution Neural Network 
-------------------


**INPUT => CONV + RELU => POOL => CONV + RELU => POOL => FC + RELU => FC => OUTPUT**

# SUMMARY OF LeNet

Nhn xt: 
* y l 1 kin trc mng CNN n gin ra i nm 1994 bi Yann LeCun 
* 3 Convolution Layers
    * s dng hm activation: sigmoid hoc tanh dn n **vanishing gradient** 
    * no padding
* 2 Pooling Layers xen gia vi stride = 2.
* 2 Fully-connected Layers v s dng hm action: softmax  d on u ra. ( Bng trn thiu 1 FC Layer )
* 1 Output Layer c 10 nodes.


MNG LeNet5  ci thin hn so vi LeNet1, ACCURACCY trn b data MNIST t 0.99% khi dng Lenet5. 

V th, Lenet5  ph bin hn v khi ni n LeNet th ni ti kin trc mng LeNet5.
